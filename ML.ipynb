```python
def evaluation_matrix(actual, pred):
  MAE = mean_absolute_error(actual, pred)
  MSE = mean_squared_error(actual, pred)
  RMSE = np.sqrt(mean_squared_error(actual, pred))
  SCORE = r2_score(actual, pred)
  return print("r2_score:",SCORE , "\n","MAE:", MAE, "\n","MSE",MSE, "\n","RMSE", RMSE)
```

```python
def evaluate_model_cv(model, X, y, cv=10):
    # Perform cross-validation and calculate the mean score
    cv_scores = cross_val_score(estimator=model, X=X, y=y, cv=cv)
    mean_cv_score = cv_scores.mean()
    return mean_cv_score
```

```python
# Ensure all target values are positive by adding a constant
min_y_train = y_train.min()
min_y_test = y_test.min()
shift_constant = 1 - min(min_y_train, min_y_test) if min(min_y_train, min_y_test) <= 0 else 0

y_train_shifted = y_train + shift_constant
y_test_shifted = y_test + shift_constant

# Ignore warnings to avoid interruption by specific models
warnings.filterwarnings('ignore')

# LazyRegressor
reg = LazyRegressor(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = reg.fit(X_train, X_test, y_train_shifted, y_test_shifted)

# Display model performance
print(models)
```
